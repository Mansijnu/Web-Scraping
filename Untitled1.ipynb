```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
r=requests.get("https://www.jnu.ac.in/se")
```

```python
c=r.content
```

```python
soup=BeautifulSoup(c,"html.parser")
```

```python
print(soup)
```

```python
images=soup.find_all('img')
```

```python
print(images)
```

```python
title=soup.title
```

```python
print(title)
```

```python
text=soup.find_all('p')
```

```python
for i in text:
    print(i.text)
```

```python
links=soup.find_all('link')
```

```python
print(links)
```
